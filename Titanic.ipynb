```python
x=df_Titanic[['Pclass']]
y=df_Titanic[['Survived']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred=regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print('r2 score corresponding to random state: ', r_state,'is: ',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print('max r2 score corresponding to',final_r_state,'is',max_r_score)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=20,scoring='r2').mean()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=31,test_size=0.20)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
#cross validaion
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
print('r2 score is ',r2_score(y_test,y_pred))
print('MSE is',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.externals import joblib
joblib.dump(lr,'lr.Titanic.csv')
```
